### Variables
Variables are fundamental elements in programming used to store data that can be referenced and manipulated in a program. In Python, variables are created when you assign a value to them, and they do not need explicit declaration to reserve memory space. The declaration happens automatically when you assign a value to a variable.

```python
## Declaring And Assigning Variables

age=25
height=5.11
name="Nikhil"
is_student=False

## printing the variables

print("age :",age)
print("Height :",height)
print("Name :",name)

if (is_student):
    print(name,"is a student")
else:
    print(name,"is not a student")
```

```python
## Naming Conventions
'''
Variable names should be descriptive
They must start with a letter or an '_' and contains letter,numbers and underscores in between
variables names case sensitive
'''

#valid variable names

first_name="Nikhil"
last_name="Kathait"
```

```python
# Invalid variable names
#2age=25
#first-name="Nikhil"
#@name="Nikhil"
```

```python
## case sensitivity
name="Nikhil"
Name="Kathait"

## Both are different variables and prints different values
print(name)
print(Name)
```

```python
## Understnading Variable types
## Python is dynamically typed, type of a variable is determined at runtime
age=25 #int
height=5.11 #float
name="Nikhil" #str
is_student=True #bool

print(type(is_student))
```

```python
## Dynamic Typing
## Python allows the type of a variable to change as the program executes
variable=10 #int
print(variable,type(variable))

variable="Hello" #str
print(variable,type(variable))

variable=3.14 #float
print(variable,type(variable))

```

```python
### Simple calculator
num1 = float(input("Enter first number: ")) #15
num2 = float(input("Enter second number: ")) #25

sum = num1 + num2
difference = num1 - num2
product = num1 * num2
quotient = num1 / num2

print("Sum:", sum)
print("Difference:", difference)
print("Product:", product)
print("Quotient:", quotient)
```
